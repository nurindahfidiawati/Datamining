1.	Masukkan iris.csv ke dataframe

```python
df = pd.read_csv('iris.csv')
df
```

2.	Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max sehingga setiap nilai attribut memiliki range 0 – 1. Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Vbaru = nilai data di suatu sel setelah normalisasi, Vlama = nilai data asli sebelum normalisasi, Vmax = nilai max dari kolom data tersebut, Vmin = nilai min dari kolom data tersebut       

```python
def normalisasi(atribut):
  return (df[atribut] - df[atribut].min()) / (df[atribut].max() - df[atribut].min())
```

```python
df["sepal.length"] = normalisasi("sepal.length")
df["sepal.width"] = normalisasi("sepal.width")
df["petal.length"] = normalisasi("petal.length")
df["petal.width"] = normalisasi("petal.width")
df
```

3.	Cari centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata2 seluruh data suatu kolom

```python
cen_setosa = df.loc[df['variety'] == 'Setosa'].mean()
cen_versicolor = df.loc[df['variety'] == 'Versicolor'].mean()
cen_virginica = df.loc[df['variety'] == 'Virginica'].mean()

print("Centroid Setosa:\n{} \nCentroid Versicolor:\n{} \nCentroid Virginica:\n{}".format(cen_setosa, cen_versicolor, cen_virginica))
```

5.	Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance 

```python

import pandas as pd
import numpy as np
df = pd.read_csv('iris.csv')
dftesting = pd.read_csv('iristesting.csv')
attribut= ['sepal.length','sepal.width','petal.length','petal.width']

def normalisasi(atribut):
  return (df[atribut] - df[atribut].min()) / (df[atribut].max() - df[atribut].min())

df["sepal.length"] = normalisasi("sepal.length")
df["sepal.width"] = normalisasi("sepal.width")
df["petal.length"] = normalisasi("petal.length")
df["petal.width"] = normalisasi("petal.width")


dflama = pd.read_csv('iris.csv')

def normalisasi(atribut):
  return (dftesting[atribut] - dflama[atribut].min()) / (dflama[atribut].max() - dflama[atribut].min())

dftesting["sepal.length"] = normalisasi("sepal.length")
dftesting["sepal.width"] = normalisasi("sepal.width")
dftesting["petal.length"] = normalisasi("petal.length")
dftesting["petal.width"] = normalisasi("petal.width")


def euclidian(datake,dataframe):
  x=['Setosa','Versicolor','Virginica']
  b=[]
  hasil = None
  for i in x:
    kelas = df[df['variety'] == i]
    centroid = kelas.mean()
    test = dataframe.iloc[datake,0:4]
    a = ((centroid[0]-test[0])**2+(centroid[1]-test[1])**2+(centroid[2]-test[2])**2+(centroid[3]-test[3])**2)*1/2
    b.append(a)
  if (min(b)== b[0]):
    hasil = ('Setosa')
    return hasil
  elif (min(b)== b[1]):
    hasil = ('Versicolor')
    return hasil
  else :
     hasil =('Virginica')
     return hasil

temp =[]
for j in dftesting.index:
  temp.append(euclidian (j,dftesting))
print(dftesting['variety'] == temp)
print(dftesting['variety'])

```
