2.	Jelaskan proses penyusunan dataset

**Proses penyusunan data set sebagai berikut :**

**a)	Aggregation**

Menggabung beberapa attribut menjadi sebuah attribut
Misal : untuk meneliti penyakit db, attribut suhu badan dicatat per 2 jam. Bisa diaggregasikan menjadi suhu harian

**b)	Sampling**

Sampling : Mengambil sebagian data yang bisa mewakili seluruh populasi. Jenis-jenis sampling
-	Simpel random sampling (acak)
-	Sampling without replacement (data yang telah disampling tidak boleh disampling lagi)
-	Sampling with replacement (data yang telah disampling boleh disampling lagi)
-	Stratified sampling (data dibagi menjadi beberapa bagian. Kemudian dari tiap bagian disampling secaraacak)

**c)	Dimensionality Reduction**

Makin banyak attribut, makin sulit menganalisa data, untuk menangani curse of dimensionality dilakukan dimensionality reduction : mengurangi jumlah attribut data yang digunakan. Hanya menggunakan sebagian attribut yang penting

**d)	Feature subset selection**

Proses pemilihan subset dari fitur yang relevan (variabel, prediktor) untuk digunakan dalam konstruksi model. Fitur dianggap relevan bila nilainya bervariasi secara sistematis dengan keanggotaan kategori

**e)	Feature creation**

Membuat attibut baru yang mungkin penting untuk analisa data. Misal dari jumlah penduduk bisa diturunkan menjadi pertumbuhan penduduk per tahun

**f)	Discretization and Binarization**

Discretization adalah transformasi atribut kontinyu kedalam bentuk atribut katagorikal. Sedangkan, Binarization adalah transformasi atribut kontinyu maupun diskrit kedalam bentuk atribut biner.

**g)	Attribute Transformation**

Sebuah fungsi yang memetakan seluruh rangkaian nilai dari atribut tertentu ke satu set nilai pengganti baru sehingga setiap nilai lama dapat diidentifikasi dengan salah satu nilai baru
